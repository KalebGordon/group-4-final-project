```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Read CSV
emissions_path = Path("emissions_table.csv")
emissions_df = pd.read_csv(emissions_path)
```

```python
# Check Work
emissions_df
```

```python
emissions_df = emissions_df.drop(columns='unit')
```

```python
emissions_df = emissions_df.loc[emissions_df['item'] == 'All sectors with LULUCF']
```

```python
emissions_df = emissions_df.drop(columns='item')
emissions_df = emissions_df.drop(columns='emissions_key')
```

```python
emissions_df
```

```python
emissions_df = emissions_df.drop_duplicates(keep='first')
emissions_df
```

```python
years = [c for c in emissions_df if c.startswith('year_')]
```

```python
emissions_pivot = emissions_df.melt(id_vars=['area', 'emission_type'], value_vars=years, var_name='year', value_name = 'emissions(kilotons)')
```

```python
emissions_pivot = emissions_pivot.pivot_table(index=['area', 'year'], columns='emission_type', values='emissions(kilotons)').reset_index()
emissions_pivot
```

```python
emissions_pivot['year'] = emissions_pivot['year'].str.replace("year_", "")
emissions_pivot['year'] = emissions_pivot['year'].astype(int)
```

```python
emissions_pivot.dtypes
```

```python
emissions_pivot
```

```python
emissions_pivot = emissions_pivot.rename(columns={'Emissions (CH4)':'Emissions_CH4', 'Emissions (CO2)':'Emissions_CO2', 'Emissions (CO2eq) (AR5)': 'Emissions_CO2eq_AR5', \
    'Emissions (CO2eq) from CH4 (AR5)': 'Emissions_CO2eq_from_CH4_AR5', 'Emissions (CO2eq) from F-gases (AR5)': 'Emissions_CO2eq_from_F-gases_AR5',\
        'Emissions (CO2eq) from N2O (AR5)':'Emissions_CO2eq_from_N2O_AR5', 'Emissions (N2O)':'Emissions_N2O'})
```

```python
popyears = [c for c in population_df if c.startswith('20')]
```

```python
population_pivot = population_df.melt(id_vars=['area'], value_vars=popyears, var_name='year', value_name='population')
population_pivot
```

```python
emissions_areas = list(emissions_pivot['area'].unique())
```

```python
pop_areas = list(population_df['area'].unique())
```

```python
compared = set(emissions_areas).intersection(pop_areas)
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(china_df.year, china_df.Emissions_CH4)
plt.xticks(np.arange(2000, 2021, 1), rotation=90)
plt.title('china emissions')
plt.xlabel('Years')
plt.ylabel('CH4 Emissions (kilotons)')
```

```python
X = china_df.year.values.reshape(-1,1)
```

```python
y = china_df.Emissions_CH4
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X,y)
plt.plot(X, y_pred, color='red')
plt.xticks(np.arange(2000, 2021, 1), rotation=90)
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
