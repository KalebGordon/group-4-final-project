```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Read all CSVs
emissions_df = pd.read_csv("emissions_data.csv")
population_df = pd.read_csv("population_data.csv")
death_df = pd.read_csv('death_data.csv')
```

## Cleaning and Pivoting the emissions dataframe.

```python
# Check dataframe.
emissions_df
```

```python
# Dropping unnecessary 'unit' column. 
emissions_df = emissions_df.drop(columns='unit')
```

```python
# Filtering the item type we want 'All sectors with LULUCF'.
emissions_df = emissions_df.loc[emissions_df['item'] == 'All sectors with LULUCF']
```

```python
# Dropping unnecessary 'item' and 'emissions_key' columns. 
emissions_df = emissions_df.drop(columns='item')
emissions_df = emissions_df.drop(columns='emissions_key')
```

```python
# Checking progress.
emissions_df
```

```python
# Dropping duplicated rows and checking row count.
emissions_df = emissions_df.drop_duplicates(keep='first')
emissions_df
```

```python
# Creating a list of years for the creation of our pivot. 
years = [c for c in emissions_df if c.startswith('year_')]
```

```python
# Pivoting table to show data longitudinally. 
emissions_pivot = emissions_df.melt(id_vars=['area', 'emission_type'], value_vars=years, var_name='year', value_name = 'emissions(kilotons)')
emissions_pivot = emissions_pivot.pivot_table(index=['area', 'year'], columns='emission_type', values='emissions(kilotons)').reset_index()
emissions_pivot
```

```python
# Chaning the years to display only numbers and converting the data type to int. 
emissions_pivot['year'] = emissions_pivot['year'].str.replace("year_", "")
emissions_pivot['year'] = emissions_pivot['year'].astype(int)
```

```python
# Checking dtypes.
emissions_pivot.dtypes
```

```python
emissions_pivot
```

```python
# Renaming columns to allow for easier fetching.
emissions_pivot = emissions_pivot.rename(columns={'Emissions (CH4)':'Emissions_CH4', 'Emissions (CO2)':'Emissions_CO2', 'Emissions (CO2eq) (AR5)': 'Emissions_CO2eq_AR5', \
    'Emissions (CO2eq) from CH4 (AR5)': 'Emissions_CO2eq_from_CH4_AR5', 'Emissions (CO2eq) from F-gases (AR5)': 'Emissions_CO2eq_from_F-gases_AR5',\
        'Emissions (CO2eq) from N2O (AR5)':'Emissions_CO2eq_from_N2O_AR5', 'Emissions (N2O)':'Emissions_N2O'})
```

```python
# Checking work.
emissions_pivot
```

```python
# Checking number of unique areas, as this will be our primary key when merging other dataframes, along with the year. 
emissions_pivot['area'].nunique()
```

## Cleaning and pivoting the population dataframe.

```python
population_df.head()
```

```python
# Creating a list of years to become a column.
popyears = [c for c in population_df if c.startswith('y2')]
```

```python
# Renaming 'Country Name' to 'area' for merging purposes.
population_df = population_df.rename(columns = {'Country_Name': 'area'})
```

```python
# Creating population pivot table.
population_pivot = population_df.melt(id_vars=['area'], value_vars=popyears, var_name='year', value_name='population')
population_pivot
```

```python
population_pivot['year'] = population_pivot['year'].str.replace("y", "")
population_pivot
```

## Merging everything together.

```python
# Making a list of every unique area in the dataframes, then creating a list of areas that they share.
emissions_areas = list(emissions_pivot['area'].unique())
pop_areas = list(population_df['area'].unique())
death_areas = list(death_df['area'].unique())
compared = set(emissions_areas).intersection(pop_areas)
compared = set(emissions_areas).intersection(death_areas)
```

```python
# There are 158 unique countries shared amongst all data sets.
len(compared)
```

```python
# Removing any area not in the compared list of areas.
emissions_pivot = emissions_pivot[emissions_pivot['area'].isin(compared)]
emissions_pivot['area'].nunique()
```

```python
# Fixing some type errors (pt1)
emissions_pivot.dtypes
```

```python
# Fixing some type errors (pt2)
population_pivot.dtypes
population_pivot['year'] = population_pivot['year'].astype(int)
population_pivot.dtypes
```

```python
# Merging emissions and population.
emissions_dataframe = pd.merge(emissions_pivot, population_pivot, on=['area', 'year'])
```

```python
# Merging death and emissions.
emissions_dataframe = pd.merge(emissions_dataframe, death_df, on=['area', 'year'])
```

```python
# Checking work. All Good!
emissions_dataframe.reset_index()
emissions_dataframe
```

```python
# Saving pivot tables.
emissions_pivot.to_csv('emissions_cleaned.csv')
population_pivot.to_csv('population_cleaned.csv')
death_df.to_csv('death_cleaned.csv')
```
